## Program 14: Prime Check

**Problem**: Read an integer `n` and print `Prime` or `Not Prime`.

**Input**:
- One integer `n`

**Output**:
- `Prime` or `Not Prime`

**Sample Input**:
```
29
```
**Sample Output**:
```
Prime
```


````python
import math


def check_prime(n):
    if n<=1:
        return("Not Prime")

    elif n>=2:
        sqrt_n=math.floor(n**(1/2))

        for i in range(2,sqrt_n+1):
            if n%i==0:
                return("Not Prime")
                
    return("Prime")


n=int(input("Enter Number: "))
# n=3
res=check_prime(n)
print(res)
````

## Program 28: Second Largest (Distinct)

**Problem**: Read `N` and a list of integers. Print the second largest **distinct** number.

**If second largest does not exist**, print: `Not Possible`.

**Input**:
- Integer `N`
- Next line: `N` integers

**Output**:
- Second largest distinct value or `Not Possible`

**Sample Input**:
```
5
5 1 5 3 2
```
**Sample Output**:
```
3
```


````python

#Find second largest number
n=int(input("Enter Length of List: "))
nums=list(map(int,input(f"Enter space seperated {n} Integers: ").split()))
# n=7
# nums=[0, 1, 2, 4, 4, 3, 2]


nums.sort()
max=nums[-1]

c = nums.count(max)
for i in range(c):
    nums.remove(max)

if len(nums) == 0:
    print("Not Possible")
else:
    print(nums[-1])
````

## Program 29: Rotate List Right by K

**Problem**: Read `N`, list of `N` integers, and `K`. Rotate the list to the right by `K` positions.

**Input**:
- Integer `N`
- Next line: `N` integers
- Integer `K`

**Output**:
- Rotated list

**Sample Input**:
```
5
1 2 3 4 5
2
```
**Sample Output**:
```
4 5 1 2 3
```


````python
n = int(input("Enter Length of List: "))

if n == 0:
    print("List is empty")
else:
    nums = list(map(int, input(f"Enter space separated {n} Integers: ").split()))
    k = int(input("Rotate right by: "))

    k = k % n   # Handle k > n
    # print(k)
    if k == 0:
        print(" ".join(map(str, nums)))
    else:
        last_k = nums[-k:]
        rem = nums[:-k]
        res = last_k + rem
        print(" ".join(map(str, res)))
````
